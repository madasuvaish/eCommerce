```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the datasets
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")

# Merge the datasets
transactions = transactions.merge(products, on="ProductID", how="left")
transactions = transactions.merge(customers, on="CustomerID", how="left")

# Step 1: Find the most sold product
product_sales = transactions.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False)
most_sold_product = product_sales.idxmax()
most_sold_quantity = product_sales.max()

# Step 2: Analyze customers who bought the most of this product
most_sold_product_transactions = transactions[transactions['ProductName'] == most_sold_product]
customer_purchase_details = most_sold_product_transactions.groupby('CustomerName')['Quantity'].sum().sort_values(ascending=False)

# Step 3: Transaction details of the most sold product
transaction_details = most_sold_product_transactions[['TransactionID', 'CustomerName', 'Quantity', 'TransactionDate', 'TotalValue']]

# Display insights
print(f"\nThe product sold the most is '{most_sold_product}' with a total quantity of {most_sold_quantity}.")

print("\nCustomers who bought the most of this product:")
print(customer_purchase_details)

print("\nTransaction details for the most sold product:")
print(transaction_details.head())  # Displaying top 5 transactions

# --- Statistical Insights and Visualization ---

# 1. Statistical Summary of Product Sales
sales_summary = product_sales.describe()
print("\nStatistical Summary of Product Sales:")
print(sales_summary)

# 2. Visualizing the Quantity of Products Sold (Bar Plot)
plt.figure(figsize=(12, 6))
sns.barplot(x=product_sales.index, y=product_sales.values, palette='viridis')
plt.title("Fig.4:Total Quantity Sold for Each Product", fontsize=16)
plt.xlabel("Product", fontsize=12)
plt.ylabel("Quantity Sold", fontsize=12)
plt.xticks(rotation=90)
plt.show()

# 3. Distribution of Quantities Purchased by Customers (Most Sold Product)
plt.figure(figsize=(10, 6))
sns.histplot(most_sold_product_transactions['Quantity'], bins=20, kde=True, color='skyblue')
plt.title(f"Distribution of Quantities Purchased for '{most_sold_product}'", fontsize=16)
plt.xlabel("Quantity Purchased", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# 4. Top Customers' Contribution (Pie Chart)
top_customers = customer_purchase_details.head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_customers, labels=top_customers.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3", n_colors=10))
plt.title(f"Top 10 Customers Who Bought Most of '{most_sold_product}'", fontsize=16)
plt.show()

# 5. Transaction Value Over Time (Line Plot)
most_sold_product_transactions['TransactionDate'] = pd.to_datetime(most_sold_product_transactions['TransactionDate'])
transactions_per_date = most_sold_product_transactions.groupby('TransactionDate')['TotalValue'].sum()

plt.figure(figsize=(12, 6))
transactions_per_date.plot(kind='line', marker='o', color='green', linewidth=2)
plt.title(f"Total Transaction Value Over Time for '{most_sold_product}'", fontsize=16)
plt.xlabel("Transaction Date", fontsize=12)
plt.ylabel("Total Transaction Value (USD)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the datasets
customers = pd.read_csv("Customers.csv")
products = pd.read_csv("Products.csv")
transactions = pd.read_csv("Transactions.csv")

# Merge the datasets
final_data = transactions.merge(products, on="ProductID", how="left")
final_data = final_data.merge(customers, on="CustomerID", how="left")

# Now you can use final_data for analysis
product_demand = final_data.groupby('ProductName')['Quantity'].sum()

# Find the least and most purchased products
least_purchased = product_demand.idxmin()
most_purchased = product_demand.idxmax()

# Data for the pie chart (top 5 highest + least purchased product for contrast)
top_demand = product_demand.sort_values(ascending=False).head(5)
top_demand[least_purchased] = product_demand[least_purchased]  # Include the least purchased

# Pie chart visualization
plt.figure(figsize=(8, 8))
top_demand.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False, explode=[0.1 if p == most_purchased else 0 for p in top_demand.index])
plt.title('Fig:6 Product Demand: Highest and Least Purchased Products')
plt.ylabel('')  # Remove default ylabel for better aesthetics
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'SignupDate' and 'TransactionDate' are in datetime format
final_data['SignupDate'] = pd.to_datetime(final_data['SignupDate'])
final_data['TransactionDate'] = pd.to_datetime(final_data['TransactionDate'])

# Calculate ConversionTime (in days)
final_data['ConversionTime'] = (final_data['TransactionDate'] - final_data['SignupDate']).dt.days

# Scatter plot for SignupDate vs ConversionTime
plt.figure(figsize=(10,6))
plt.scatter(final_data['SignupDate'], final_data['ConversionTime'], alpha=0.5, color='blue')
plt.title('Fig:6 Customer Conversion Time Analysis')
plt.xlabel('Signup Date')
plt.ylabel('Conversion Time (Days)')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by Region and sum up the TotalValue to identify sales by region
region_sales = final_data.groupby('Region')['TotalValue'].sum().reset_index()

# Sort by sales volume
region_sales = region_sales.sort_values(by='TotalValue', ascending=False)

# Plot the total sales by region
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='TotalValue', data=region_sales, hue='Region', palette='viridis', legend=False)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales Value (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
